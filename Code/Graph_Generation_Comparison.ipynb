```python
import matplotlib.pyplot as plt
import pandas as pd

# Define model specs
models = [
    {"name": "3CNN-1DC", "macc": 40.3, "acc": 80.00, "weights": 51.24, "ram": 103.13, "latency": 595.14},
    {"name": "3CNN-1DC Distilled", "macc": 40.3, "acc": 81.54, "weights": 51.24, "ram": 103.13, "latency": 595.15},
    {"name": "5CNN-1DC", "macc": 67.2, "acc": 80.00, "weights": 60.11, "ram": 113.13, "latency": 992.03},
    {"name": "6CNN-1DC", "macc": 80.6, "acc": 88.46, "weights": 63.68, "ram": 118.13, "latency": 1190.74},
    {"name": "2CNN-1DC Distilled", "macc": 15.1, "acc": 79.23, "weights": 46.80, "ram": 56.88, "latency": 550.61},
    {"name": "2CNN-1DC Distilled (RAM Optimized)", "macc": 15.1, "acc": 79.23, "weights": 46.80, "ram": 52.39, "latency": 653.83}
]

# Create DataFrame
df = pd.DataFrame(models)

# Plot Accuracy vs Weights
fig1, ax1 = plt.subplots()
for i, row in df.iterrows():
    ax1.scatter(row["weights"], row["acc"], label=row["name"])
ax1.set_title("Model Accuracy vs Weights (KiB)")
ax1.set_xlabel("Weights (KiB)")
ax1.set_ylabel("Accuracy (%)")
ax1.grid(True)
ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Plot Accuracy vs RAM
fig2, ax2 = plt.subplots()
for i, row in df.iterrows():
    ax2.scatter(row["ram"], row["acc"], label=row["name"])
ax2.set_title("Model Accuracy vs RAM (KiB)")
ax2.set_xlabel("RAM (KiB)")
ax2.set_ylabel("Accuracy (%)")
ax2.grid(True)
ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Plot Latency vs MACCs
fig3, ax3 = plt.subplots()
for i, row in df.iterrows():
    ax3.scatter(row["latency"], row["macc"], label=row["name"])
ax3.set_title("Latency vs MACCs")
ax3.set_xlabel("Latency (ms)")
ax3.set_ylabel("MACCs (Million)")
ax3.grid(True)
ax3.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
